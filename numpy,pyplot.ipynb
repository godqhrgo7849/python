### 문제
- subplot을 이용해서 1행 2열 구조 2개씩
- 4가지 색상 다르게
- 4가지 스타일 다르게
- 4가지 마커 서로 다르게
- 범례표시
- x,y,제목

```python
x=np.arange(0,6,1)
y1= x+1
y2= x+2
y3= x+3
y4= x+4
```

```python
rc('font',family='gulim')
plt.subplot(1,2,1)
plt.plot(y1,'r',marker='*')
plt.plot(y2,'g--',marker='+')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title1')
plt.legend('RG')
plt.subplot(1,2,2)
plt.plot(y3,'b',marker='.')
plt.plot(y4,'y',marker='^')
plt.xlabel('엑스')
plt.ylabel('와이')
plt.title('제목')
plt.legend('BY')
```
