## 주제: 경찰청-가정폭력 검거 및 조치현황

## 출처 : https://www.data.go.kr/data/15037060/fileData.do

- 분류체계   :공공질서및안전 - 경찰
- 제공기관   :경찰청
- 조사된 기간:2015~2017년

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
path = r'C:\Users\user\Desktop\303\data.csv'
path
```

```python
data=pd.read_csv(path,encoding='cp949')
data
```

```python
l=data['연도']
x=data['검거건수(건)']
x1=data['검거인원총계(명)']
x2=data['구속인원']
x3=data['불구속인원']
```

```python
rc('font',family='gulim')
plt.plot(l,x,'r--',marker='*',label='검거건수')
plt.plot(l,x1,'b--',marker='o',label='검거인원총계(명)')
plt.xlabel('연도')
plt.ylabel('건수')
plt.title('가정폭력')
plt.legend()
plt.show()
```

- 2015년부터 2016년도까지 가정폭력 검거건수가 점차 증가하면서 4만5천권 이상인것을 볼 수 있다.
- 2017년도에는 약6천권이 감소한 것을 볼 수 있다.
- 검거인원는 검거건수을 보면 2016년이 많듯 2016년에 인원이 가장 많으며 평균적으로 약4만8천명인 것을 볼 수 있다.
- 그리고 가정폭역으로 피해를 보는 사람들이 많은 것을 알 수 있다.

```python
plt.subplot(1,2,1)
plt.tight_layout()
plt.plot(l,x2,'b--',label='구속인원(명)')
plt.legend()
plt.xlabel('연도')
plt.ylabel('명')
plt.title('가정폭력')
plt.subplot(1,2,2)
plt.tight_layout()
plt.plot(l,x3,'y--',label='불구속인원(명)')
plt.legend()
plt.xlabel('연도')
plt.ylabel('명')
plt.title('가정폭력')
```

- 년도별로 구속인원과 불구속인원이 줄어드는 것을 볼 수 있다.
- 불구속인원이 4만명 이상인 것이 보이며, 2016년도에 5만명 이상인 것을 볼 수 있다.
- 불구속인원이 많은 만큼 다시 가정폭력에 시달리거나 혹은 정신적으로 피해를 입는 사람이 생길 수 있다는 생각이 든다.
