```python
import matplotlib.pyplot as plt
a=[10,-20, 30, 40]
b=['kor','eng', 'math','info']
plt.bar(b,a,color='y')
plt.ylabel('jumsu')
plt.xlabel('subject')
plt.title('SCORE')
plt.legend('value')
plt.show()
```

```python
import matplotlib.pyplot as plt
a=[10,-20, 30, 40]
b=['kor','eng', 'math','info']
c=[1,2,3,4]
plt.bar(a,c,color='y')
plt.fill(a,b,color='b')
plt.ylabel('jumsu')
plt.xlabel('subject')
plt.title('SCORE')
plt.legend('value')
plt.show()
```
