```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
img=r'C:\Users\user\Desktop\303\B.png'
```

```python
img
```

```python
Image.open(img)
```

```python
lp_mask = np.array(Image.open(img))
lp_mask
```

```python
tx=open(r'C:\Users\user\Desktop\303\A.txt',encoding='utf-8').read()
```

```python
tx
```

```python
lp_wc=WordCloud(
    font_path = r'C:\Windows\Fonts\gulim.ttc',
    background_color = 'white',
    mask=lp_mask,
    max_font_size=50,
    contour_width=5,
    contour_color='green').generate(tx)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file(r'C:\Users\user\Desktop\303\B.png')
```
